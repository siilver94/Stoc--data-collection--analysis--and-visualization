```python
import seaborn as sns

# 만들어둔 krx.csv 파일을 pd.read_csv 로 불러옵니다.
df = pd.read_csv("krx.csv",dtype={"Symbol":object})
#Code를 Symbol로 바꿔줍니다
df = df.rename(columns={'Code':'Symbol'})

# head 로 위에 있는 데이터 일부만 가져옵니다.
df.head(3)
```

```python
# Market의 빈도수를 구합니다.
# 상위 30개 Market의 빈도수를 market_count_top 변수에 할당합니다.
market_count_top = df["Market"].value_counts().head(30)
market_count_top.index

df_market_30  = df[df["Market"].isin(market_count_top.index)]

# 빈도수가 가장 많은 데이터 색인하기
plt.figure(figsize=(10,8))
sns.countplot(data=df_market_30, y="Market", palette="magma",
              order=market_count_top.index).set_title("마켓별 빈도수")
```

```python
# Departmenet의 빈도수를 구합니다.
# industry_top30

Departmenet_top30 = df["Dept"].value_counts().head(30)

plt.figure(figsize=(10,6))
sns.countplot(data=df[df["Dept"].isin(Departmenet_top30.index)],
             y="Dept", 
              order=Departmenet_top30.index).set_title("Department Count")
```

```python
# MarketID의 빈도수를 구합니다.
MarketID_count = df["MarketId"].value_counts().head(30)
MarketID_count.index

df[df["MarketId"].isin(MarketID_count.index)]

# order 기능을 통해 빈도수를 정렬합니다.
# df["MarketID"].value_counts().index 로 MarketID 의 순서를 가져와서 지정합니다.
sns.countplot(data = df[df["MarketId"].isin(MarketID_count.index)],
             y = "MarketId", order = MarketID_count.index).set_title("Market Count")
```
